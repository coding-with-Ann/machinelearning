```python
import pandas as pd 
```

```python
df = pd.read_csv('customer.csv') 
df.head()
```

```python
df['purchased'].unique()
```

```python
encode_columns = df[['review', 'education', 'purchased']]
```

```python
from sklearn.preprocessing import OrdinalEncoder 
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG'], ['No', 'Yes']])
```

```python
oe.fit(encode_columns)
```

```python
encode_columns = oe.transform(encode_columns)
```

```python
encoded_result = pd.DataFrame(encode_columns, columns= ['review', 'education', 'purchased'], dtype=int)
```

```python
encoded_result['review'].unique()
```

```python
encoded_result.head()
```

```python
cars = pd.read_csv('/home/ann/Documents/MachineLearning-main/FeatureEngineering/day27-one-hot-encoding/cars.csv')
cars.head()
```

```python
cars['brand'].unique()
```

```python
car_brands = cars[['brand']]
car_brands
```

```python
from sklearn.preprocessing import OneHotEncoder 
ohe = OneHotEncoder(categories=[['Maruti', 'Skoda', 'Honda', 'Hyundai', 'Toyota', 'Ford', 'Renault',
       'Mahindra', 'Tata', 'Chevrolet', 'Fiat', 'Datsun', 'Jeep',
       'Mercedes-Benz', 'Mitsubishi', 'Audi', 'Volkswagen', 'BMW',
       'Nissan', 'Lexus', 'Jaguar', 'Land', 'MG', 'Volvo', 'Daewoo',
       'Kia', 'Force', 'Ambassador', 'Ashok', 'Isuzu', 'Opel', 'Peugeot']], sparse_output=False)
```

```python
ohe.fit(car_brands)
```

```python
car_brands = ohe.transform(car_brands)
```

```python
car_brands_encoded = pd.DataFrame(car_brands, dtype=int)
car_brands_encoded.head()
```

```python
passengers = pd.read_csv('/home/ann/Documents/MachineLearning-main/FeatureEngineering/day32-binning-and-binarization/train.csv')
passengers.head()
```

```python
passengers['Age'].info()
```

```python
passengers_age = passengers['Age'].dropna()
passengers_age.isna()
```

```python
passengers_age = pd.DataFrame(passengers_age)
```

```python
from sklearn.preprocessing import KBinsDiscretizer 

bins = KBinsDiscretizer(n_bins=3, encode='onehot', strategy='quantile')
```

```python
bins.fit(passengers_age)
```

```python
passengers_age_transformed = bins.transform(passengers_age)
```
