### pipelines

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('/home/ann/Documents/machinelearning/FeatureEngineering/day29-sklearn-pipelines/train.csv')
```

```python
data.head()
```

```python
x = data[['Pclass', 'Sex', 'Age', 'Fare']]
```

```python
y = data['Survived']
```

```python
x.head()
```

```python
y.head()
```

```python
x.isnull().sum()
```

```python
y.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
trf1 = ColumnTransformer([
    ('simple_impute', SimpleImputer(), [2])
], remainder='passthrough')
```

```python
trf2 = ColumnTransformer([
    ('sex', OneHotEncoder(sparse_output=False), [1]), 
     ('age', KBinsDiscretizer(n_bins=10, strategy='quantile', encode='onehot'), [2]),
     ('fare', KBinsDiscretizer(n_bins=15, strategy='quantile', encode='ordinal'), [3])
], remainder='passthrough')
```

```python
trf3 = ColumnTransformer([
    ('standard_scale', StandardScaler(), [0,2,3])
], remainder='passthrough')
```

```python
pl = Pipeline([
    ('missing_values', trf1),
    ('encode', trf2),
    ('scale', trf3)
])
```

```python
pl.fit(x_train)
```
